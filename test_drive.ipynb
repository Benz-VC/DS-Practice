# Step 4
Model Training Train Test

```python
## Import sklearn train-test-split model

from sklearn.model_selection import train_test_split as tts
```

```python
## Transform data linear regression model variables (with existing features)

X = data.iloc[:, [5, 11, 12]]
y = data.iloc[:, 2]

print('X Variables')
print(X.head(5))
print('y Variable')
print(y.head(5))
```

```python
## Train-test-split model

X_train, X_test, y_train, y_test = tts(X, 
                                       y,
                                       test_size = 0.2,
                                       random_state = 42
                                       )

## Check output
print('X Train Output')
print(X_train.head(10))

print('y Train Output')
print(y_train.head(10))
```

```python
## Linear Regression Model Training
## Import linear regression library

from sklearn.linear_model import LinearRegression as LR

## Prompt of success import lib

print('Linear Regression Model Import Complete')
```

```python
## Train model 

reg = LR().fit(X_train, y_train)

## Prompt of complete training
print('Training Complete')
```

```python
## Prediction

y_pred = reg.predict(X_test)

## Prompt
print('Prediction Complete')
```

# Step 5
Data analysis and visualization

```python
## Import measurement libraries

from sklearn.metrics import mean_squared_error, r2_score

## Key stats
## Coefficeint
print('Coefficients: \n', reg.coef_)
## Mean Squared Error (MSE)
print('Mean squared error: %.2f' % mean_squared_error(y_test, y_pred))
## Coefficient of determination
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))

## Summary

print(reg.n_features_in_)
```

```python
## Summary 

## Import more complex statistics libraries 
import statsmodels.api as sm

## Data preparation
X_train_ar = np.array(X_train)
y_train_ar = np.array(y_train)

## Model fit
model = sm.OLS(y_train_ar, X_train_ar)
print('Training Complete')

## Key Result Display
results = model.fit()

print(results.summary())
```

```python
## Residual plot

plt.scatter(y_test, y_pred)
```

```python
## Error Distribution

plt.hist(y_test - y_pred)
```

```python
## Calculate Price per Sqm
## Lot
price_per_sqm_lot = data.iloc[:, 2] / data.iloc[:, 6]
print('Price per Sqm (Lot)')
print(price_per_sqm_lot.head(10))

## Living Area
price_per_sqm_living = data.iloc[:, 2] / data.iloc[:, 5]
print('Price per Sqm (Living)')
print(price_per_sqm_living.head(10))
```

```python
## Simple Linear Regression Test

## Transform data linear regression model variables (with existing features)

X = data.iloc[:, [5]]
y = data.iloc[:, 2]

print('X Variables')
print(X.head(5))
print('y Variable')
print(y.head(5))

## Train-test-split model

X_train, X_test, y_train, y_test = tts(X, 
                                       y,
                                       test_size = 0.1,
                                       random_state = 0
                                       )

## Model training

reg = LR().fit(X_train, y_train)
print('Training Complete')

## Prediction

y_pred = reg.predict(X_test)

## Prompt
print('Prediction Complete')
```

```python
## Data Analysis

## Import measurement libraries

from sklearn.metrics import mean_squared_error, r2_score

## Key stats
## Coefficeint
print('Coefficients: \n', reg.coef_)
## Intercept
print('Y Intercept: \n', reg.intercept_)
## Mean Squared Error (MSE)
print('Mean squared error: %.2f' % mean_squared_error(y_test, y_pred))
## Coefficient of determination
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))

## Number of Features

print('Number of Features: %.2f' % reg.n_features_in_)
```
